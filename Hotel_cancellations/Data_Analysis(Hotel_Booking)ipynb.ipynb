#Data analysis and Visualizations

```python
cancelled_prec=df['is_canceled'].value_counts(normalize=True)
cancelled_prec.name = 'is_canceled'
print(cancelled_prec)
```

```python
counts = df['is_canceled'].value_counts()

# Define the categories and their labels
labels = ['Not canceled', 'Canceled']

# Ensure that counts are ordered according to the labels
# '0' corresponds to 'Not canceled' and '1' corresponds to 'Canceled'
values = [counts.get(0, 0), counts.get(1, 0)]

# Create the bar plot
plt.figure(figsize=(5, 5))
plt.title('Reservation Status Count')
plt.bar(labels, values, edgecolor='black', width=0.7)
plt.xlabel('Reservation Status')
plt.ylabel('Proportion')
plt.show()
```

The bar graph shows the percentage of reservations that are cancelled and those that not. It is obvious are still a significant no of reservations are not cancelled. There are still 37% of clients who cancelled their reservation, which has a significant impact on the hotels earnings.

```python
plt.figure(figsize=(10,5))
ax1=sns.countplot(x='hotel',hue='is_canceled',data=df,palette='Blues')
legend_labels,_=ax1.get_legend_handles_labels()
plt.legend(bbox_to_anchor=(1.05,1))
plt.title('Reservation Status Count')
plt.xlabel('Hotel')
plt.ylabel('number of reservations')
```

In comparisions to resort hotels, city hotels have more bookings. It's possble that resort hotels are more expensive thant those cities.

```python
resort_hotel=df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel=df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel=resort_hotel.groupby('reservation_status_date')['adr'].mean()
city_hotel=city_hotel.groupby('reservation_status_date')['adr'].mean()
```

```python
plt.figure(figsize=(15,8))
plt.title('Average Daily Rate in City and Resort Hotel',fontsize=30)
plt.plot(resort_hotel.index,resort_hotel,label='Resort Hotel')
plt.plot(city_hotel.index,city_hotel,label='city Hotel')
plt.legend(fontsize=30)
plt.show()
```

The line graph shows that on certain days, the average daily rate for a city hotel is less than that of a resort hotel, and on other days, it is even less. It goes without saying that weekends and holidays may see a less rise in resort hotel rates.

```python
df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(10,5))
ax1=sns.countplot(x='month',hue='is_canceled',data=df,palette='Blues')
```

The grouped bar graph is done to analyze the months with the highest and lowest reservation levels according to reservation status. Both the number of confirmed reservations and cancelled reservations are largest in the month of August.

```python
plt.figure(figsize=(10,5))
plt.title('AQR per month')
sns.barplot(x='month',y='adr',data=df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index(),palette='viridis')
plt.legend(fontsize=20)
plt.show()
```

The bar graph demonstrates that cancellations are most common when prices are greatest and least when they are lowest. Therefore, the cost of the accomodation is solely responsible for cancellation.

```python
cancelled_data=df[df['is_canceled']==1]
top_10_country=cancelled_data['country'].value_counts()[:10]
plt.figure(figsize=(8,5))
plt.title('Top 10 countries with reservation cancellations')
plt.pie(top_10_country,labels=top_10_country.index,autopct='%1.1f%%')
plt.show()
```

The area from where guests are coming and making reservations. Around 46% of the clients come from online travel agencies, whereas 27% comes from groups. Only 4% cleints book hotels by visiting them.

```python
# For cancelled data
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')['adr'].mean().reset_index()
cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

# For not cancelled data
not_cancelled_data=df[df['is_canceled']==0]
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')['adr'].mean().reset_index()
not_cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

# Plotting the data
plt.figure(figsize=(15,8))
plt.title('Average Daily Rate', fontsize=20)
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='Cancelled')
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='Not Cancelled')
plt.legend(fontsize=15)
plt.show()
```

Reservations are cancelled when the average daily rate is higher then when it is not cancelled.

```python
cancelled_df_adr=cancelled_df_adr[(cancelled_df_adr['reservation_status_date']>'2016')&(cancelled_df_adr['reservation_status_date']<'2017-09')]
not_cancelled_df_adr=not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date']>'2016')&(not_cancelled_df_adr['reservation_status_date']<'2017-09')]
```

```python
plt.figure(figsize=(15,8))
plt.title('Average Daily Rate', fontsize=20)
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='Not Cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='Cancelled')
plt.legend(fontsize=15)
```
